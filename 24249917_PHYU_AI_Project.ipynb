# ðŸ”¹ Step 5: Demographic Recommender

The implementation of demographic-based recommendation is relatively straightforward. It involves ranking movies according to their overall ratings and presenting the highest-rated titles to users. The process typically includes the following steps:

(1) establishing a metric to evaluate or assign a score to each movie

(2) computing this score for all available movies, and

(3) sorting the results to identify and recommend the top-rated selections to users.

The Weighted Rating (WR) can be expressed as:

Weighted Rating (WR) = (v Ã· (v + m)) Ã— R + (m Ã· (v + m)) Ã— C


where:

v represents the number of votes a movie has received,

m denotes the minimum number of votes required for inclusion in the ranking,

R is the average rating of the movie, and

C indicates the mean rating across all movies in the dataset.

Given that the dataset already includes the variables v (vote_count) and R (vote_average), the next step is to compute C, the overall mean rating.

```python
movies['vote_average'].mean()
```


The subsequent step involves identifying a suitable threshold for m, which represents the minimum number of votes a movie must receive to qualify for inclusion in the ranking.
In this case, 90% is used as the cutoff point which mean only movies with a vote count higher than 90% of all listed titles are considered for the chart.

```python
movies['vote_count'].quantile(q=0.9) # 0.9 -> 90%
df_movies_filtered=movies[movies['vote_count']>movies['vote_count'].quantile(q=0.9)]
df_movies_filtered.shape
```

There are 481 movies which qualify to be in this list.
To calculate score for each qualified movie, we can define a function, weighted_rating(), and apply this function to the DataFrame movies filter of qualified movies.

```python
def weighted_rating(x):
    v=x['vote_count']
    m=movies['vote_count'].quantile(q=0.9)
    R=x['vote_average']
    C=movies['vote_average'].mean()
    return ((R*v)/(v+m))+((C*m)/(v+m))
```

```python
print(weighted_rating)
movies['score']=movies.apply(weighted_rating, axis=1)
```

# ðŸ”¹ Step 7 : Content-Based Recommender

Defining the Recommendation Function -> find_similar_movies

The final step is to create a function that takes a movie title as input and returns the top 10 most similar movie recommendations.


1. Get Movie Index: Given a movie title, the function must first locate its corresponding index in the original DataFrame.

2. Retrieve similarity scores => sim_scores = list(enumerate(cosine_sim[idx]))


- cosine_sim is an nÃ—n matrix containing similarity scores between every pair of
movies.

- cosine_sim[idx] retrieves the similarity vector for the selected movie (its similarity with every other movie).

- enumerate() pairs each similarity score with the corresponding movie index, producing a list of tuples
matrix.

3. Sort by similarity

4. Select top 10 similar movies

5. Extract indices of similar movies

6. Retrieve movie titles

Get Movie Index : Title & index

```python
movies_indices=pd.Series(data=list(movies.index), index= movies['title'] )
movies_indices.head()
```

The find_similar_movies() function retrieves the most semantically related movies by leveraging cosine similarity scores computed from feature vectors generated through CountVectorizer. The function ranks all movies based on their similarity to the selected title, filters the top-10 highest scores, and outputs the corresponding movie titles as recommendations.

```python
# input -> movie tile, output -> similar movies
def find_similar_movies(title, cosine_sim):

    # Get the index of the movie that matches the title
    idx = movies_indices[title]

    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores.sort(key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores=sim_scores[1:11]

    # Get the movie indices
    ind=[]
    for (x,y) in sim_scores:
        ind.append(x)

    # Return the top 10 most similar movies
    tit=[]
    for x in ind:
        tit.append(movies.iloc[x]['title'])
    return pd.Series(data=tit, index=ind)
```

```python
# Extract the director's name from the crew feature. If none found, return NaN
def get_director(crew_list):
    if isinstance(crew_list, list):  # ensure input is a list
        for member in crew_list:
            if member.get('job') == 'Director':
                return member.get('name')
    return None  # returning None is better than the string 'NaN'
```

```python
# Extract up to the top 3 names from a list of dictionaries (e.g., cast, genres, keywords)
def extract_top3_names(item_list):
    if isinstance(item_list, list):
        return [item.get('name') for item in item_list[:3]]
    return []
```

# ðŸ”¹ Step 15: Collaborative Recommender System

The content-based engine was a solid first step, but it suffers from a few key limitations that prevent it from providing truly excellent recommendations:

1.	"Filter Bubble" Problem: Only capable of suggesting movies that are extremely close to the input movie which it struggles to capture diverse tastes or provide recommendations across different genres which tends to keep suggesting the same type of content.

2.	Zero Personalization: The recommendations aren't personal at all. Anyone who queries the engine using the same input movie will get the same exact list, regardless of their unique viewing history or preferences.
To move beyond this and achieve genuine personalization, new approach will be applied.

**Introducing Collaborative Filtering (CF)**

Collaborative Filtering (CF) is a powerful technique that achieves personalization without relying on movie metadata (like genres or actors). The core principle is simple: CF matches users with similar interests.
Instead of analysing the content of the movies, it analyses the pattern of user ratings to build relationships.

*Implementing CF with the Surprise Library*

To build and analyse this new recommender system, the Surprise library can be used. Surprise is a dedicated Python package for recommender systems that deals explicitly with rating data.

Within Surprise, one of the most powerful algorithms for CF: Singular Value Decomposition (SVD) can be employed.

â€¢	SVD is a matrix factorization technique used to discover the latent (hidden) factors that explain the relationship between users and items.

â€¢	To create the most accurate predictions possible, which means minimizing the error. We will measure this error using the Root Mean Square Error (RMSE).

â€¢	To ensure the new model is robust and not just overfitting to the training
data, we will rigorously evaluate its performance using K-Fold Cross-Validation.

By leveraging SVD and proper validation can create a truly personal and effective recommender system!

# ðŸ”¹ Step 15: Hybrid Recommender

Input: User ID and the Title of a Movie

Output: Similar movies sorted on the basis of expected ratings by particular user

```python
# Get the current column names
current_columns = movies.columns.tolist()
print(f"Current number of columns: {len(current_columns)}")
print(f"Current columns: {current_columns}")

# The original list had 25 names, but the DataFrame has 27.
# Let's inspect the DataFrame head to see the column names and adjust accordingly.
# Based on the previous outputs, the extra columns are 'score' and 'metadata'.
# I will add these to the list of column names.
movies.columns=['budget', 'genres', 'homepage', 'id', 'keywords', 'original_language',
       'original_title', 'overview', 'popularity', 'production_companies',
       'production_countries', 'release_date', 'revenue', 'runtime',
       'spoken_languages', 'status', 'tagline', 'title', 'vote_average',
       'vote_count', 'movie_id', 'cast', 'crew', 'director', 'actor', 'score', 'metadata']
movies.head(2)
```

```python
def generate_personalized_recommendations(user_id, movie_title):
  #get movie index
  idx= movies_indices[movie_title]
  # get pair similarity scores of all movies with user selected movie
  sim_scores= list(enumerate(cosine_sim[idx]))

  # sort movie with similarity scores
  sim_scores.sort(key=lambda x: x[1], reverse=True)
  # get top 10 similar movies
  sim_scores=sim_scores[1:11]

  # get movie indices
  ind=[]
  for (x,y) in sim_scores:
      ind.append(x)
  # get title, movieid, vote_average , vote-count for top 10 similar movies
  title = []
  moviesID = []
  vote_average = []
  vote_count = []
  for x in ind:
    title.append(movies.iloc[x]['title'])
    moviesID.append(movies.iloc[x]['movie_id'])
    vote_average.append(movies.iloc[x]['vote_average'])
    vote_count.append(movies.iloc[x]['vote_count'])

    # rating prediction for a prticular user for top 10 similar movies

    rate_predit = []
    for a in moviesID:
      rate_predit.append(svd.predict(user_id, a).est)

  return pd.DataFrame({'index': ind, 'title':title, 'movieId':moviesID, 'vote_average':vote_average, 'vote_count':vote_count,'estimated_rating':rate_predit}).set_index('index').sort_values(by='estimated_rating', ascending=False)
```

```python
# get personalized recommandation for particular user
generate_personalized_recommendations(1, 'Avatar')
```
